### Given two polynomial numbers represented by a linked list. The task is to complete the function addPolynomial() that adds these lists meaning adds the coefficients who have the same variable powers.

```python
class node:
    def __init__(self, coeff, pwr):
        self.coef = coeff
        self.next = None
        self.power = pwr
        
# Linked List Class
class Linked_List:
    def __init__(self):
        self.head = None

    def insert(self, coeff, pwr):
        if self.head == None:
            self.head = node(coeff, pwr)
        else:
            new_node = node(coeff, pwr)
            temp = self.head
            while(temp.next):
                temp=temp.next
            temp.next = new_node

def createList(arr, n):
    lis = Linked_List()
    k=0
    for i in range(n):
        lis.insert(arr[k], arr[k+1])
        k+=2
    return lis.head

'''
 class node:
    def __init__(self, coeff, pwr):
        self.coef = coeff
        self.next = None
        self.power = pwr
'''

class Solution:
    # return a linked list denoting the sum with decreasing value of power
    def addPolynomial(self, poly1, poly2):
        # Code here
        head = node(0,0)
        t = head
        while poly1 != None and poly2 != None:
            if poly1.power == poly2.power:
                t.next = node(poly1.coef + poly2.coef, poly1.power)
                poly1 = poly1.next
                poly2 = poly2.next
                
            elif poly1.power > poly2.power:
                t.next = node(poly1.coef, poly1.power)
                poly1 = poly1.next
                
            elif poly2.power > poly1.power:
                t.next = node(poly2.coef, poly2.power)
                poly2 = poly2.next
            t = t.next
        while poly1 != None:
            t.next = poly1    
            t = t.next
            poly1 = poly1.next
        while poly2 != None:
            t.next = poly2
            t = t.next
            poly2 = poly2.next
        return head.next

if __name__=='__main__':
    t = int(input())
    for i in range(t):
        n = int(input())
        arr = list(map(int, input().strip().split()))
        poly1 = createList(arr, n)
        n = int(input())
        arr = list(map(int, input().strip().split()))
        poly2 = createList(arr, n)
        sum = Solution().addPolynomial(poly1, poly2)
        ptr = sum
        while ptr is not None:
            print(str(ptr.coef) + 'x^' + str(ptr.power), end = '')
            ptr = ptr.next
            if ptr is not None:
                print(' +', end = ' ')
        print()
```
